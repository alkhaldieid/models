```
unclean = []
for i in range(np.shape(unnumFeatures92)[0]):
  for j in range(np.shape(unnumFeatures92)[1]):
    if isinstance(unnumFeatures92[i,j], str):
      unclean.append([i,j])
print(unclean)
len(unclean)
```

```
uncleanRows=[]
for i in range(len(unclean)):
  uncleanRows.append(unclean[i][0])
print(len(uncleanRows))
print(uncleanRows)
```

```
cleanFeatures92 = np.delete(unnumFeatures92, uncleanRows, axis= 0)
print(np.shape(cleanFeatures92))
```

```
for i in range(len(cleanFeatures92)):
  print((cleanFeatures92[i,:]))
```

```
unclean_labels = unnumData92[:,10]
print(len(unclean_labels))
cleanLabels92 = np.delete(unclean_labels, uncleanRows, axis=0)
print(len(cleanLabels92))
```

```
import keras
categorical_labels=keras.utils.to_categorical(cleanLabels92)
```

```
training_data = np.array(list(zip(cleanFeatures92,categorical_labels)))
print(np.shape(training_data))
```

```
np.shape(training_data)
```

```
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras import metrics
```

```
#simple keras model
t_model = Sequential()
t_model.add(Dense(100, activation="relu"))
t_model.add(Dense(50, activation="relu"))
t_model.add(Dense(10))
t_model.compile(
    loss="mean_squared_error",
    optimizer=Adam(lr=0.001),
    metrics=[metrics.mae])
```

```
epochs = 500
batch = 128
```

```
training_data
```

```
history = t_model.fit(
    x = cleanFeatures92,
    y = categorical_labels,
    batch_size=batch,
    epochs=epochs,
    shuffle=True,
    verbose=1,
    callbacks=[]
)
```
